```python
grouped = df.groupby('ID')
grouped['impact force (mN)'].std()
```

```python
def coeff_of_var(data):
    """Compute coefficient of variation from an array of data."""
    return np.std(data) / np.mean(data)
```

```python
df3 = grouped[ ['impact force (mN)','adhesive force (mN)'] ].agg(coeff_of_var).reset_index()
df3
```

```python
def mean_(data):
    return np.mean(data)
```

```python
stat = (
    df.groupby("ID")[["impact force (mN)", "adhesive force (mN)"]]
    .agg([np.mean, np.median, np.std, coeff_of_var])
    .reset_index()
)
stat
```

```python
#tidy this list 
stat_tidy = pd.melt(stat,value_name = 'value', var_name = ['force','statistic'],id_vars = 'ID')
stat_tidy
```
