```python
import pandas as pd
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
df = pd.read_csv("data/gfmt_sleep.csv",na_values = '*')
df['insomnia'] = df['sci']<16
df.head()
```

```python
x = 'confidence when correct'
y = 'confidence when incorrect'

p = bokeh.plotting.figure(
    frame_width = 400,
    frame_height = 300,
    x_axis_label = x,
    y_axis_label = y
)
```

```python
#create the glyphs as circle
p.circle(
    source = df,
    x = x , #name of the df column of interest
    y = y ,
)
```

```python
bokeh.io.show(p)
```

```python
#distinguish insomniac  and non insomniac
# For convenience
x = 'confidence when correct'
y = 'confidence when incorrect'

# Make figure
p = bokeh.plotting.figure(
    frame_width=400,
    frame_height=300,
    x_axis_label=x,
    y_axis_label=y,
    tooltips = [
        ('p-number', '@{participant number}'), #braces because there is a space
        ('gender', '@gender'),
        ('age', '@age'),
    ]
)

# Add glyphs
p.circle(
    source=df.loc[~df['insomnia'], :],
    x=x,
    y=y,
    legend_label='normal sleepers'
)

p.circle(
    source=df.loc[df['insomnia'], :],
    x=x,
    y=y,
    color='orange',
    legend_label='insomniacs'
)


p.legend.location = 'top_left'
p.legend.click_policy = "hide"

bokeh.io.show(p)
```

```python
bokeh.io.save(
    p,
    filename = "insomniac_confidence_correct.html",
    title = 'Sample Bokeh plot'             
)
```

```python
import numpy as np
import pandas as pd

import iqplot

import bokeh.plotting
import bokeh.io

bokeh.io.output_notebook()
```

```python
df = pd.read_csv('data/frog_tongue_adhesion.csv',comment = '#')
df.head()
```

```python
p = iqplot.box(
    data = df,
    q = "impact force (mN)",
    cats = 'ID'
) 
```

```python
bokeh.io.show(p)
```

```python
p = iqplot.strip(
    data=df,
    q="impact force (mN)",
    cats="ID",
)

bokeh.io.show(p)
```

```python
p = iqplot.stripbox(
    data=df,
    q="impact force (mN)",
    cats="ID",
)

bokeh.io.show(p)
```

```python
p = iqplot.histogram(
    data=df,
    q="impact force (mN)",
    cats="ID",
)

bokeh.io.show(p)
```

```python
# Generate normally distributed data
rg = np.random.default_rng(3252)
x = rg.normal(size=500)

# Plot the histogram
p = iqplot.histogram(x, rug=False)

bokeh.io.show(p)
```

```python
p = iqplot.ecdf(x)

bokeh.io.show(p)
```

```python
p = iqplot.ecdf(
    data=df,
    q="impact force (mN)",
    cats="ID",
)

bokeh.io.show(p)
```

```python
p = iqplot.ecdf(
    data=df,
    q="impact force (mN)",
    cats="ID",
    style='staircase',
)

bokeh.io.show(p)
```
